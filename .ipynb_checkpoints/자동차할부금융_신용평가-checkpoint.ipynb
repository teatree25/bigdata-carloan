#### Ìá¥ÏßÅÏûê

```python
import pandas as pd

# üëâ Î®ºÏ†Ä ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î≥µÏÇ¨ (Ïó¨Í∏∞ÏÑ† df_retiredÍ∞Ä ÏõêÎ≥∏)
df_retired = df_retired.copy()  # Ïà´Ïûê Îß§ÌïëÎßå Ï†ÅÏö©
df_retired_dummy = df_retired.copy()  # Ïà´Ïûê Îß§Ìïë + ÎçîÎØ∏Ìôî

# -----------------------------------
# Ïà´Ïûê Îß§Ìïë (Îëê Í∞úÏùò Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î™®ÎëêÏóê Ï†ÅÏö©)
# -----------------------------------

education_map = {
    'ÎåÄÌïôÏõê': 1, 'ÎåÄÌïôÍµê': 2, 'Ï†ÑÎ¨∏ÎåÄ': 3, 'Í≥†Îì±ÌïôÍµê': 4, 'Ï§ëÌïôÍµê': 5
}
incomeclass_map = {
    '3000ÎßåÏõê ÎØ∏Îßå': 1,
    '3000ÎßåÏõê~5000ÎßåÏõê ÎØ∏Îßå': 2,
    '5000ÎßåÏõê~1ÏñµÏõê ÎØ∏Îßå': 3,
    '1ÏñµÏõê Ïù¥ÏÉÅ': 4
}
residence_map = {
    'ÏµúÏÉÅ': 1,
    'ÏñëÌò∏': 2,
    'ÌèâÍ∑†': 3,
}
week_map = {
    'Ïõî': 1, 'Ìôî': 2, 'Ïàò': 3, 'Î™©': 4, 'Í∏à': 5
}

def classify_accompany(x):
    if x in ['Î∞∞Ïö∞', 'Í∏∞ÌÉÄ', 'ÏûêÎÖÄ', 'ÏπúÏßÄ', 'Îã®Ï≤¥']:
        return 'ÎèôÌñâ'
    elif x == 'Îã®ÎèÖ':
        return 'Îã®ÎèÖ'
    else:
        return

# Ïà´Ïûê Îß§Ìïë - Îëê Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Î™®ÎëêÏóê ÎèôÏùºÌïòÍ≤å Ï†ÅÏö©
for df in [df_retired, df_retired_dummy]:
    df['Education'] = df['Education'].map(education_map)
    df['IncomeClass'] = df['IncomeClass'].map(incomeclass_map)
    df['ResidenceClass'] = df['ResidenceClass'].map(residence_map)
    df['ApplWeek'] = df['ApplWeek'].map(week_map)
    df['Accompany'] = df['Accompany'].apply(classify_accompany)

# -----------------------------------
# df_retired_dummyÎßå get_dummies Ï≤òÎ¶¨
# -----------------------------------


cols_get_dummy = [
    'Gender', 'MaritalStatus', 'IncomeType', 'OrgType',
    'HouseOwnYN', 'CarOwnedYN', 'ActiveLoanYN', 'LoanType', 'HousingType',
    'Accompany', 'InterestType', 'HomeAddMatchedYN', 'WorkAddMatchedYN',
    'Occupation', 'DwellingType', 'Education','IncomeClass','ResidenceClass'
]

df_retired_dummy = pd.get_dummies(df_retired_dummy, columns=cols_get_dummy, drop_first=False)
```

## Ïã†Ïö©Ï†êÏàò Í≥ÑÏÇ∞

```python
weights = {
    'EmployedYears': 0.2094,
    'Age': 0.0953,
    'ResidenceClass_3': 0.0906,
    'Education_2': 0.0845,
    'PhoneChangedYears': 0.0754,
    'IdChangedYears': 0.0509,
    'InfoChangedYears': 0.0447,
    'InterestRate': 0.0421,
    'InquiryCount': 0.0415,
    'Occupation_ÏòÅÏóÖ': 0.0345,
    'Occupation_ÏöîÎ¶¨': 0.0271,
    'Occupation_ÎÖ∏Îèô': 0.0226,
    'IncomeType_ÏÑúÎπÑÏä§': 0.0193,
    'Occupation_Î≥¥Ïïà': 0.0170,
    'OrgType_ÎÜçÏóÖ': 0.0156,
    'OrgType_Î¨¥Ïó≠': 0.0148,
    'FamilyCount': 0.0136,
    'Occupation_ÏùºÏö©ÏßÅ': 0.0123,
    'MaritalStatus_Í≤∞Ìòº': 0.0122,
    'Education_4': 0.0119
}
```

```python
# Ïã†Ïö©Ï†êÏàò Í≥ÑÏÇ∞ Ìï®Ïàò
from sklearn.preprocessing import MinMaxScaler

def calculate_credit_score(df, weights, score_col='Score_Re'):
    df = df.copy()

    # Ï†êÏàò Í≥ÑÏÇ∞ (Í∞ÄÏ§ëÏπò * Î≥ÄÏàò Í∞íÏùò Ìï©)
    df[score_col] = sum(df.get(var, 0) * weight for var, weight in weights.items())

    # Ï†ïÍ∑úÌôî (0~1)
    scaler = MinMaxScaler()
    df[score_col + '_scaled'] = scaler.fit_transform(df[[score_col]])

    return df
```

```python
# Ïã†Ïö© Ï†êÏàò Í≥ÑÏÇ∞
dft_emp_scored = calculate_credit_score(dft_emp, weights)
dft_retired_scored = calculate_credit_score(dft_retired, weights)

# Í≤∞Í≥º ÌôïÏù∏
print('Ïû¨ÏßÅÏûê Ïã†Ïö©Ï†êÏàò')
print(dft_emp_scored[['Score_Re', 'Score_Re_scaled']].head())
print('Ìá¥ÏßÅÏûê Ïã†Ïö©Ï†êÏàò')
print(dft_retired_scored[['Score_Re', 'Score_Re_scaled']].head())
```

### Ïã†Ïö©Ï†êÏàò Î≥ÄÌôòÍ≥º Îì±Í∏â Íµ¨Í∞Ñ ÏÑ§Ï†ï

```python
def assign_credit_grade(df, score_col='Score_Re_scaled'):
    df = df.copy()

    # 1. Ïã†Ïö©Ï†êÏàò: 0~1 Ï†ïÍ∑úÌôîÎêú Ï†êÏàòÎ•º 0~1000 Î≤îÏúÑÎ°ú Î≥ÄÌôò
    df['CreditScore'] = (df[score_col] * 1000).round(0).astype(int)

    # 2. Ïã†Ïö©Îì±Í∏â Î∂ÄÏó¨
    def grade(score):
        if score <= 453:
            return 'ÏúÑÌóò'
        elif score <= 629:
            return 'Ï£ºÏùò'
        elif score <= 767:
            return 'ÏùºÎ∞ò'
        else:
            return 'Ïö∞Îüâ'

    df['CreditGrade'] = df['CreditScore'].apply(grade)
    
    return df
```

```python
# dft_emp_scored ÎòêÎäî dft_retired_scored Ïóê Ï†ÅÏö©
dft_emp_final = assign_credit_grade(dft_emp_scored)
dft_retired_final = assign_credit_grade(dft_retired_scored)
```

```python
# Ïã†Ïö©Îì±Í∏â Î∂ÑÌè¨ Î≥¥Í∏∞
print(dft_emp_final['CreditGrade'].value_counts())
print(dft_retired_final['CreditGrade'].value_counts())

# ÏùºÎ∂Ä ÏÉòÌîå Ï∂úÎ†•
print(dft_emp_final[['CreditScore', 'CreditGrade']].head())
```
