## cf) 범주형 데이터 -> 더미화한 변수랑 안한거랑 나누기
- 매핑 + (더미화 X) : df_retired, df_emp
- 매핑 + 더미화 : df_retired_dummy, df_emp_dummy

#### 퇴직자

```python
import pandas as pd

# 👉 먼저 원본 데이터프레임 복사 (여기선 df_retired가 원본)
df_retired = df_retired.copy()  # 숫자 매핑만 적용
df_retired_dummy = df_retired.copy()  # 숫자 매핑 + 더미화

# -----------------------------------
# 숫자 매핑 (두 개의 데이터프레임 모두에 적용)
# -----------------------------------

education_map = {
    '대학원': 1, '대학교': 2, '전문대': 3, '고등학교': 4, '중학교': 5
}
incomeclass_map = {
    '3000만원 미만': 1,
    '3000만원~5000만원 미만': 2,
    '5000만원~1억원 미만': 3,
    '1억원 이상': 4
}
residence_map = {
    '최상': 1,
    '양호': 2,
    '평균': 3,
}
week_map = {
    '월': 1, '화': 2, '수': 3, '목': 4, '금': 5
}

def classify_accompany(x):
    if x in ['배우', '기타', '자녀', '친지', '단체']:
        return '동행'
    elif x == '단독':
        return '단독'
    else:
        return

# 숫자 매핑 - 두 데이터프레임 모두에 동일하게 적용
for df in [df_retired, df_retired_dummy]:
    df['Education'] = df['Education'].map(education_map)
    df['IncomeClass'] = df['IncomeClass'].map(incomeclass_map)
    df['ResidenceClass'] = df['ResidenceClass'].map(residence_map)
    df['ApplWeek'] = df['ApplWeek'].map(week_map)
    df['Accompany'] = df['Accompany'].apply(classify_accompany)

# -----------------------------------
# df_retired_dummy만 get_dummies 처리
# -----------------------------------


cols_get_dummy = [
    'Gender', 'MaritalStatus', 'IncomeType', 'OrgType',
    'HouseOwnYN', 'CarOwnedYN', 'ActiveLoanYN', 'LoanType', 'HousingType',
    'Accompany', 'InterestType', 'HomeAddMatchedYN', 'WorkAddMatchedYN',
    'Occupation', 'DwellingType', 'Education','IncomeClass','ResidenceClass'
]

df_retired_dummy = pd.get_dummies(df_retired_dummy, columns=cols_get_dummy, drop_first=False)
```

```python
# 신용점수 계산 함수
from sklearn.preprocessing import MinMaxScaler

def calculate_credit_score(df, weights, score_col='Score_Re'):
    df = df.copy()

    # 점수 계산 (가중치 * 변수 값의 합)
    df[score_col] = sum(df.get(var, 0) * weight for var, weight in weights.items())

    # 정규화 (0~1)
    scaler = MinMaxScaler()
    df[score_col + '_scaled'] = scaler.fit_transform(df[[score_col]])

    return df
```

### 신용점수 변환과 등급 구간 설정

```python
def assign_credit_grade(df, score_col='Score_Re_scaled'):
    df = df.copy()

    # 1. 신용점수: 0~1 정규화된 점수를 0~1000 범위로 변환
    df['CreditScore'] = (df[score_col] * 1000).round(0).astype(int)

    # 2. 신용등급 부여
    def grade(score):
        if score <= 453:
            return '위험'
        elif score <= 629:
            return '주의'
        elif score <= 767:
            return '일반'
        else:
            return '우량'

    df['CreditGrade'] = df['CreditScore'].apply(grade)
    
    return df
```

```python
# dft_emp_scored 또는 dft_retired_scored 에 적용
dft_emp_final = assign_credit_grade(dft_emp_scored)
dft_retired_final = assign_credit_grade(dft_retired_scored)
```

```python
# 신용등급 분포 보기
print(dft_emp_final['CreditGrade'].value_counts())
print(dft_retired_final['CreditGrade'].value_counts())

# 일부 샘플 출력
print(dft_emp_final[['CreditScore', 'CreditGrade']].head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# 신용등급별 신용점수 분포 히스토그램
sns.histplot(data=dft_emp_final, x='CreditScore', hue='CreditGrade',
             multiple='stack', bins=40, palette='Set2', edgecolor='black')

plt.title('신용등급별 신용점수 분포 (Histogram)', fontsize=14)
plt.xlabel('신용점수 (0~1000)')
plt.ylabel('인원수')
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
